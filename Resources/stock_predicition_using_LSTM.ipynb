# Data Preprocessing

```python
#splitting data into training and testing set
data_training=pd.DataFrame(df.Close[0:int(len(df)*0.7)])
data_testing=pd.DataFrame(df.Close[int(len(df)*0.7):int(len(df))])
```

```python
data_testing
```

```python
data_training
```

# scaling down between 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array=scaler.fit_transform(data_training)
data_training_array
```

```python
type(data_training_array)
```

```python
data_training_array.shape
```

# Sequence Creation

```python
#previous day price depend
x_tran=[]
y_tran=[]
for i in range(100,data_training_array.shape[0]):
    x_tran.append(data_training_array[i-100:i])
    y_tran.append(data_training_array[i,0])
x_tran,y_tran=np.array(x_tran),np.array(y_tran)
```

```python
x_tran
```

```python
x_tran.ndim
```

```python
y_tran
```

```python
y_tran.ndim
```

```python
y_tran.shape
```

# Model Building

#keras is high level api building and creating model

```python
import sys
print(sys.executable)
```

```python
import tensorflow as tf
print(tf.__version__)
print('ggg')
```

```python
# create sequential model at keraas
from keras.models import Sequential
from keras.layers import Dense,LSTM,Input
```

```python
# sequential model is used LSTM we use
model=Sequential()
model.add(Input(shape=(100,1)))
model.add(LSTM(units=128,activation='tanh',return_sequences=True))
model.add(LSTM(units=64))
model.add(Dense(25))
model.add(Dense(1))
```

# Model Training

```python
#here epoch reading the data just fifity times
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_tran,y_tran,epochs=50)
```

```python
model.summary()
```

```python
model.save('stock-prediction-model.keras')
```
