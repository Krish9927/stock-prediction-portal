## SubPlots And Legends

```python
day=[1,2,3,4,5]
orignal_price=[10,20,30,40,50]
predicited_price=[8,21,23,21,32]
```

```python
plt.plot(day,orignal_price,'b-o',label='original price')
plt.plot(day,predicited_price,'r-o',label='predicited price')
plt.legend()
plt.title("Plot with Legends")
plt.xlabel('days')
plt.ylabel('Predicited price')
```

#subplot

```python
plt.subplot(3,1,1)
plt.plot(day,orignal_price,'b-o',label='original price')
plt.title("Plot with Legends")
plt.xlabel('days')
plt.ylabel('Predicited price')
plt.legend()


plt.subplot(3,1,2)

plt.plot(day,predicited_price,'r-o',label='predicited price')
plt.title("Plot with Legends days")
plt.xlabel('days jm')
plt.ylabel('Predicikmted price')
plt.legend()

```
